```python
#Working function
def check_for_date(d):
    tw1 = pd.read_csv(d + '/{}-dataset.tsv'.format(d))
    tw2 = pd.read_csv(d + '/{}_clean-dataset.tsv'.format(d))
    tw3 = pd.read_csv(d + '/{}_top1000bigrams.csv'.format(d))
    tw4 = pd.read_csv(d + '/{}_top1000terms.csv'.format(d),  header=None)
    tw4.columns = ['term', 'count']
    tw5 = pd.read_csv(d + '/{}_top1000trigrams.csv'.format(d))
    return tw1, tw2,tw3, tw4, tw5
```

```python
def check_for_date_top_1000terms(d):
    tw4 = pd.read_csv(d + '/{}_top1000terms.csv'.format(d),  header=None)
    tw4.columns = ['term', 'count']
    tw4['term'] = tw4['term'].astype(str)
    tw4 =  tw4[tw4['term'].map(lambda x: x.isascii())]
    return tw4
```

```python
def check_for_date_top1000bigrams(d):
    tw3 = pd.read_csv(d + '/{}_top1000bigrams.csv'.format(d))
    tw3.columns = ['term', 'count']
    return tw3
```
