# Loading and storing data in MongoDB Cloud Cluster

```python
# need the following to connect to MongoDB cloud cluster
import pymongo
import dns

# need the following to read JSON files in a directory
import json
import os
```

**1) Connect to MongoDB cloud cluster**

```python
client = pymongo.MongoClient("__your_mongodb_cloud_cluster_url_here__")
```

```python
**2) Insert data into MongoDB cloud cluster**
```

```python
db_name = "nytimes_raw"  # set database name to create in mongodb cluster
directory = 'json_data/nytimes_raw' # set directory that contains all the raw json data

db = client[db_name]  # create/connect to database

for file in os.listdir(directory):
    
    # create/connect to collection in database
    collection = db[file]

    # read JSON file (as dict type object)
    f = open('json_data/nytimes_raw/' + file, 'r')
    data = json.loads(f.read())
    
    # insert data into collection
    collection.insert_many(data)
```

### Check database status

```python
# list existing databases in mongodb cluster
for x in client.list_database_names():
    print(x)
```

```python
# list existing collections in database
db = client['nytimes_raw']

print(f'There are a total of {len(db.list_collection_names())} collections. \n')

for collname in db.list_collection_names():
    print(collname)
```
